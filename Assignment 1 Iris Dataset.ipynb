```python
# Importing the libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Task 1. Read the dataset to python environment.

```python
# Reading the data from the file.
data = pd.read_excel(r'C:\Users\sreeh\DSA\Data-sets\iris.xls',sheet_name="Sheet1")
```

#### Task 2. Display the columns in the dataset.

```python
# Reading the columns, assigning it to a variable and then dispalying it.
column_disp=list(data.columns)
print("The columns are ",column_disp)
```

#### Task 3. Calculate the mean of each column of the dataset.

```python
# Calculating the mean of all the columns and displaying it
mean1=data.mean(numeric_only=True)
print("Mean of each columns is ")
mean1
```

#### Task 4. Check for the null values present in the dataset.

```python
# Checking if there are any null values and adding it up.
print("The sum total of null values in the set is \n")
data.isna().sum()
```

#### Task 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
# First let us do a simple bar chart and compare how all three species compare.
data.groupby("Classification")[['SL','SW','PL','PW']].sum().plot(kind='barh')
plt.show()
```

```python
# Second, let us use a scatter plot to compare the Petal Length and Petal width for each species.
sns.scatterplot(x='PL',y='PW', hue='Classification',data=data,)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

```python
# Third, we can use a box plot to see how the data is arranged.
plt.figure(figsize=(7,7))
sns.boxplot(x=data['Classification'],
            y=data["PL"])
plt.ylabel("Petal Length")
plt.show()
```

```python
# Now let us see a histogram
sns.histplot(data,x="SL",hue='Classification',kde=True)
plt.xlabel("Sepal Length")
plt.show()
```

```python
# Finally a correlation heat map.
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True,cmap="PiYG")
plt.show()
```
